```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
waa = pd.read_csv("waa.csv")
grades = pd.read_csv("grades.csv")
receiving = pd.read_csv("receiving.csv")
```

```python
temp = pd.merge( left = grades, right = waa, how = 'left', on = ['player_id', 'season'])
```

```python
temp = temp.drop(columns=['player_first_name', 'player_last_name'])
```

```python
temp = temp.dropna()
```

```python
temp.corr(numeric_only=True)
```

```python
heat = temp.drop(columns=['position','player','team'])
```

```python
sb.heatmap(heat, annot=True, cmap="coolwarm")
```

```python
temp.head()
```

```python
sorted_temp = temp.sort_values(by=['player', 'year_in_league'], ascending=[True, True])
sorted_temp.head(10)
```

```python
before_transfer = []
after_transfer = []
current_player = ""
current_team = ""
transferred = False
for i, row in sorted_temp.iterrows():
    if(row["player"] == current_player and row["team"] == current_team and transferred == False):
        before_transfer.append(row)
    elif(row["player"] == current_player and row["team"] == current_team and transferred == True):
        after_transfer.append(row)
    elif(row["player"] != current_player):
        current_player = row["player"]
        current_team = row["team"]
        before_transfer.append(row)
        transferred = False
    else:
        after_transfer.append(row)
        transferred = True
        current_team = row["team"]
```

```python
before_transfer = pd.DataFrame(before_transfer)
before_transfer.head()
```

```python
after_transfer = pd.DataFrame(after_transfer)
after_transfer.head()
```

```python
before_transfer[['player', 'offense', 'receiving', 'waa']].groupby('player').mean()
```

```python
team_waa = after_transfer[["team", "waa"]].groupby("team").mean()
team_waa.plot(kind="box")
plt.title("Average WAA After Transfer by Team")
plt.ylabel("WAA")
plt.xlabel("Teams")
```
